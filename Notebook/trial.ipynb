```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_col=df.columns
df_col
```

```python
plt.figure(figsize=(15, 10))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(df.columns)):
    plt.subplot(2, 2, i+1)
    sns.kdeplot(x=df[df_col[i]], color='blue')
    # Adjust layout to prevent overlap
    plt.xlabel(df_col[i])
    plt.tight_layout()
    
```

```python
sns.countplot(x="Result", data=df)
plt.show()
```

```python
df_discrete=[features for features in df_col if len(df[features].unique())<=25]
#in general, continuous variables tend to have more unique values than discrete variables.
df_continous=[features for features in df_col if len(df[features].unique())>25]
```

```python
print(f'the discrete variable are{df_discrete}')
print(f'the continous variable are{df_continous}')
```

```python
corr_matrix=df.corr()
```
